```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Распределение Лапласа

```python
x=np.linspace(-3, 3, 1000)
lap=sts.laplace()
pdf = lap.pdf(x)
plt.plot(x, pdf)
```

#### Функция распределения

```python
laplas_random=lap.rvs(size=1000)
cdf = lap.cdf(x)
plt.plot(x, cdf)
```

#### Плотность вероятности

```python
pdf = lap.pdf(x)
plt.plot(x, pdf)
```

#### Наложение графика теоретической плотности вероятности на гистограмму сгенерированного распределения Лапласа

```python
plt.hist(laplas_random,  normed=True, bins=30) #Генерация выборки
plt.plot(x, pdf)
plt.title("Распределение Лапласа")
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
```

#### Наложение граффка теоретического нормального распределения на гистограмму средних с N=5

```python
vib_sr=[]

#генерируем 1000 распределений по 5 в каждом

for i in range(1000): 
    laplas_random5=lap.rvs(size=5)
    z=(sum(laplas_random5)/5)
    vib_sr.append(z)
    
#выводим гистограмму
plt.hist(vib_sr, bins=30, normed=True)

#вычисляем mu и sigma для нормального распределения
mu = np.mean(vib_sr) 
sigma = np.std(vib_sr,ddof=1)
SE = np.sqrt(2)/np.sqrt(5)
norm_rv = sts.norm(loc=mu, scale=SE)

#Выводим полученное
norm_graf = norm_rv.pdf(x)
plt.plot(x, norm_graf)
```

#### Наложение граффка теоретического нормального распределения на гистограмму средних с N=10

```python
vib_sr=[]

for i in range(1000):
    laplas_random10=lap.rvs(size=10)
    z=(sum(laplas_random10)/10)
    vib_sr.append(z)
plt.hist(vib_sr, bins=30, normed=True)

mu = np.mean(vib_sr)
sigma = np.std(vib_sr,ddof=1)
SE = np.sqrt(2)/np.sqrt(10)
norm_rv = sts.norm(loc=mu, scale=SE)

norm_graf = norm_rv.pdf(x)
plt.plot(x, norm_graf)
```
