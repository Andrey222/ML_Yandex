```python
import numpy as np
from scipy import linalg as lnlg
from matplotlib import pylab as plt
```

```python
# Function to be approximated
def myFunc(x):
    return np.sin(x / 5.0) * np.exp(x / 10.0) + 5 * np.exp(-x / 2.0)
a = np.arange(1, 16, 1)
b = myFunc(a)
```

```python
myFunc(1)
```

# Approximation 1

```python
# Parameters of system of equations X * w = f
# where X - n*n matrix of variables; w - n*1 vector of coefficients; f - vector of function results
X1 = np.array([[1, 1], [1, 15]] )
f1 = [myFunc(1), myFunc(15)]
w1 = lnlg.solve(X1, f1)
print 'X1\n:', X1
print 'f1\n:', f1
print 'w1\n:', w1
```

```python
plt.plot(X1[:,1],f1,a,b,'r')
plt.show()
```

# Approximation 2

```python
X2 = np.array([[1, 1, 1**2], [1, 8, 8**2], [1, 15, 15**2]] )
f2 = myFunc(X2[:,1])
w2 = lnlg.solve(X2, f2)
print 'X2\n:', X2
print 'f2\n:', f2
print 'w2\n:', w2
plt.plot(X2[:,1],f2,a,b,'r')
plt.show()
```

# Approximation 3

```python
X3 = np.array([[1, 1, 1**2, 1**3], [1, 4, 4**2, 4**3],[1, 10, 10**2, 10**3], [1, 15, 15**2, 15**3]] )
f3 = myFunc(X3[:,1])
w3 = lnlg.solve(X3, f3)
print 'X3\n:', X3
print 'f3\n:', f3
print 'w3\n:', w3
plt.plot(X3[:,1],f3,a,b,'r')
plt.show()
```
